Goal: Identify largest matching subset from a string

```python
import numpy as np
import pandas as pd
```

```python
string_1 = 'Here is a string crossover that potentially has a matching subset'
```

```python
string_2 = 'This string potentially has some crossover in a subset'
```

```python
index_str = [char.lower() for char in string_1]
```

```python
col_str = [char.lower() for char in string_2]
```

```python
similarity_matrix = pd.DataFrame(columns=col_str, index=index_str, )
```

```python
def same_finder(column_index):
    vals_list = []
    for i in range(len(index_str)):
        if [char for char in similarity_matrix.index][i] == col_str[column_index]:
            vals_list.append(1)
        else:
            vals_list.append(0)
    return vals_list
```

```python
for j in range(len(col_str)):
    similarity_matrix.iloc[:, j]= same_finder(j)
```

```python
def find_subset(str1, str2):
    if str1==str2:
        return str1
    if len(str1)==0 or len(str2)==0:
        return ''
    
    if len(str1)<len(str2):
        small_str = str1
        larg_str = str2
    else:
        small_str = str2
        larg_str=str1
    min_len = len(small_str)+1
    #max_len = len(larg_str)
        
    avail_list = []
    avail_len = []
    #Ending position
    for i in range(min_len):
        #starting position
        for j in range(i):
            if small_str[j:i] in larg_str:
                #save some memory by only appendding if len greater than current len
                avail_list.append(small_str[j:i])
                avail_len.append(len(small_str[j:i]))
                
    
    return avail_list[avail_len.index(max(avail_len))]
        
        
```
